```python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image
```

```python
a_true = np.random.rand()
b_true = np.random.rand()

N = 100
x = 5 * np.random.rand(N)
y = a_true * x + b_true + np.random.normal(0,0.4,N) # np.random.normal(0,0.4,N) 은 noise => 없을때 있을때 비교해보기
```

```python
plt.plot(x,y,'.')
```

```python
def mse_loss(a, b, x, y):
    N = len(x)
    loss = 0
    for i in range(N):
        loss += np.power((y[i] - (a * x[i] + b)), 2)
    return loss
```

```python
def plot_line(a, b, x, y):
    x_range = np.arange(0, 5, 0.01)
    plt.plot(x, y, 'b.', x_range, a*x_range + b, 'r')
```

```python
print(mse_loss(a_true, b_true, x, y))
plot_line(a_true, b_true, x, y)
```

```python
a = 0.4
b = 1
print(mse_loss(a, b, x, y))
plot_line(a, b, x, y)
```

```python
ones = np.ones(N)
A_mat = np.array([[np.inner(x,x), np.inner(ones,x)], [np.inner(x,ones), N]])
b_mat = np.array([np.inner(y,x), np.inner(y, ones)])

ab_sol = np.linalg.solve(A_mat, b_mat) # https://rfriend.tistory.com/380 : 연립방적식 해 풀기
a_est, b_est = ab_sol

# 연립방적식 수식으로 표현하면 아래 사진과 같다.
Image("C://Users\이익범\Pictures\Proof.jpg") 
```

```python
print(mse_loss(a_est, b_est, x, y))
plot_line(a_est, b_est, x, y)
```

```python
print(f'a_true = {a_true}, a_estimate = {a_est}')
print(f'b_true = {b_true}, b_estimate = {b_est}')
```
